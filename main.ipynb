```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Weather Analysis\localglobal-temp.csv') #Initialize a dataframe and read CSV
```

```python
dfmean = df.rolling(window=150, center=False, on="LOCALAVGT").mean().dropna() #Set a rolling window of 150 and drop NA values
```

```python
x = dfmean.YEAR #Set x coord in dfmean (Year Values)
y = dfmean.LOCALAVGT #Set y coord in dfmeans (Delhi Temperature)
r = dfmean.GLOBALAVGT #Set r coord in dfmeans (Global temperature)
```

```python
fig = plt.figure(figsize=(7,5)) # Set the total size of the fig

plt.plot(x,r, color='blue', linewidth=2, label="Global") #Plotting the Global Temperatures against Year

fig.suptitle('Average Global Temperature(Â°C) per Year')
plt.legend() #Legend for the graph
plt.xlabel("Year", fontsize=14) 
plt.ylabel("Temperature (Â°C)", fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
```
